```python
#- Split data into Y (`total_claim_amount`) and X (numerical, independent variables)
```

```python
Y = numerical['Total Claim Amount']
X = numerical.drop(['Total Claim Amount'], axis=1)
```

```python
#Conduct Linear Regressions
#Using Statsmodels
```

```python
import numpy as np
from scipy.stats import norm
import math
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
X = sm.add_constant(X)
model = sm.OLS(Y,X).fit()

print(model.summary())
```

```python
# El modelo es 'decente' ya que R-Squared es mayor que 0.5, pero mejorable ya que es de 0,519
# Casi todos las variables son altamente significativas menos Number of Policies y Number of Open Complaints ya que tienen pval > 0,5.
# La variable de Monthly Premium Auto es muchisimo mas alto que el del resto de variables independientes, por lo que es la mas significativa.
# La interpretaciÃ³n del coeficiente es cuÃ¡nto se moverÃ­a la variable dependiente si la variable independiente sube en una unidad, manteniendo todo lo demÃ¡s constante.
```

```python
#Using sklearn
```

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

Y = numerical['Total Claim Amount']
X = numerical.drop(['Total Claim Amount'], axis=1)
lm = linear_model.LinearRegression()
model = lm.fit(X,Y)
print("R2 value is = ",round(lm.score(X,Y),2))
```

```python
predictions = lm.predict(X)
print("R2 value is = ",round(r2_score(Y, predictions),2))
print("The intercept of the model is = ",lm.intercept_)
print("The coefficients of the model are = ",lm.coef_)
predictions = lm.predict(X)
mse = mean_squared_error(Y, predictions)
print("The mse of the model is = ",mse)
```

```python
# Mismo r2 value por lo que la conclusion es la misma.
# A nivel de coeficientes vemos los mismos resultados que en statmodels.
# Sin embargo, el MSE de este modelo es muy alto lo que nos dice que tiene una probabilidad de error muy alta. Lo cual es importante minimizar.
```
